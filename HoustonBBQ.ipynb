# Load data from foursquare and rearrange 

```python
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query, radius, LIMIT)
url
```

```python
results = requests.get(url).json()
#results
```

```python
# assign relevant part of JSON to venues
venues = results['response']['venues']

# tranform venues into a dataframe
dataframe = json_normalize(venues)
dataframe.head()
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

simp_df1=dataframe_filtered.drop(['categories','labeledLatLngs','postalCode','state','country','crossStreet','neighborhood','id','cc'], axis=1)
pd.set_option('display.max_colwidth', -1)
display(simp_df1)
```

# Use Foursquare data and retrieve restaurants rating 
## If no rating is available ignore the restaurant

```python
rest_rating_list=[]
for index, row in dataframe_filtered.iterrows():
    
    venue_id=row['id']
    #print(venue_id)
    url1 = 'https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(venue_id, CLIENT_ID, CLIENT_SECRET, VERSION)
    result = requests.get(url1).json()
    result
    try:
        rest_rating=(result['response']['venue']['rating'])
        #print(rest_rating)
        rest_rating_list.append(rest_rating)
        #print(rest_rating)
    except:
        rest_rating='n/a'
        rest_rating_list.append(rest_rating)
        #rest_rating_list=rest_rating_list+rest_rating
        #print(rest_rating) 

#rest_rating_list
```

```python
dataframe_filtered['rating'] = rest_rating_list
df = dataframe_filtered[dataframe_filtered.rating != 'n/a']
#display(df)
sorted_df=df.sort_values(by='rating', ascending=False)
sorted_df.reset_index(inplace=True)
sorted_df.head()
```
