```python
# using pandas to read the html in dataframe dfs
dfs=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
len(dfs)
```

```python
# read only the first table and store in another dataframe df
df=dfs[0]
```

```python
# display the dataframe
df.head(15)
```

```python
# Drop the rows with Borough as "Not assigned"
mod_df=df[df.Borough !='Not assigned']
mod_df.head(10)
```

```python
# indetify Neighbourhood which is "No assigned"

filter = mod_df['Neighbourhood']=='Not assigned'
mod_df.where(filter).dropna()
```

```python
# Replace the Neighourhood "Not assigned" with corresponding Borough, "Queen's Park" in this case
mod_df.replace(to_replace =["Not assigned"],  
                            value ="Queen's Park") 
```

```python
# Now group the dataframe by Postcode and store in dataframe called final_df
final_df=mod_df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
final_df.head(50)
```
